<h3 style="color:#66CDAA"> Median Income By Boroughs <b>  </b></h3>

```python
#Extracting Columns
income = nyc_df_sorted['median_income']
boroughs = nyc_df_sorted.index

#Bar Plot
fig = plt.figure(figsize = (10, 8))
plt.bar(boroughs, income, color = '#66CDAA', edgecolor = '#3CB371')
plt.xticks(rotation = 0, fontsize = 13, fontweight = 'bold')
plt.yticks(rotation = 0, fontsize = 12, fontweight = 'bold')

#Labels
plt.xlabel('Boroughs', fontsize = 16)
plt.ylabel('Median Income', fontsize = 16)

#Title
plt.title('Median Income across the different Boroughs.', fontsize = 20, fontweight = 'bold')

#Show
plt.show()
```

<h3 style="color:#7B68EE"> Median Rent in the different Boroughs <b>  </b></h3>

```python
#Extracting Columns
rent = nyc_df_sorted['median_rent']

#Bar Plot
fig = plt.figure(figsize = (8, 5))

plt.barh(boroughs, rent, color = '#7B68EE', edgecolor = '#533ae9')

plt.xticks(rotation = 0, fontsize = 14, fontweight = 'bold')
plt.yticks(rotation = 0, fontweight = 'bold', fontsize = 13)

#Labels
plt.xlabel('Median Rent', fontsize = 16)
plt.ylabel('Boroughs', fontsize = 16)

#Title
plt.title('Median Rent in New York City', fontsize = 20, fontweight = 'bold')

#Show
plt.show()
```

<h3 style="color:#F17615"><b> Income | Rent Correlation: </b></h3>

```python
#Extracting Columns
income = nyc_df_sorted['median_income']

#Scatter Plot
fig = plt.figure(figsize = (10, 8))

plt.scatter(rent, income, color = '#F17615', edgecolor = '#F06B02', s = 1500, linewidths = 2)
plt.xticks(rotation = 0, fontsize = 13, fontweight = 'bold')
plt.yticks(rotation = 0, fontsize = 12, fontweight = 'bold')

#Labels
plt.xlabel('Median Rent', fontsize = 16)
plt.ylabel('Median Income', fontsize = 16)

#Title
plt.title('Income & Rent correlation across NYC boroughs.', fontsize = 20, fontweight = 'bold')

#Show
plt.show()
```

<p style="color:#F17615"><b> This is a positive correlation. This means that each of the variables is influencing the other. As x grows, so does y. </b></p>

<h5 style="color:#EE665B"><b> You would think, why is this information important?</b></h5>
<p style="color:#EE665B">It is important to know the crime rates of a borough for several reasons:
Safety: High crime rates can indicate that an area may not be safe, which can be a concern for residents and potential homebuyers. Knowing the crime rates can help individuals make informed decisions about where they want to live and can help them take appropriate precautions to stay safe.
Property value: Areas with high crime rates may have lower property values due to concerns about safety and perceived risk. This can affect the investment potential of a property, as well as the resale value.
Quality of life: Living in an area with high crime rates can have a negative impact on quality of life, leading to stress, anxiety, and a sense of unease. Knowing the crime rates can help individuals assess whether an area is a good fit for their lifestyle and preferences.
Community involvement: Understanding the crime rates can also encourage community involvement in crime prevention efforts. This can include neighborhood watch programs, community policing initiatives, and other collaborative efforts to reduce crime and improve safety.
Overall, knowing the crime rates of a borough can help individuals make informed decisions about where to live, can affect property values, quality of life, and community involvement in crime prevention efforts.</p>

<h3 style="color:#B7BA4A"><b> Unemployment | Labor Force Rate Correlation: </b></h3>

```python
#Extracting Columns
unemployment = nyc_df_sorted['unemployment']
labor = nyc_df_sorted['labor_force_rate']

#Scatter Plot
fig = plt.figure(figsize = (10, 8))

plt.scatter(labor, unemployment, color = '#B7BA4A', edgecolor = '#DBE105', s = 1500, linewidths = 2)
plt.xticks(rotation = 0, fontsize = 13, fontweight = 'bold')
plt.yticks(rotation = 0, fontsize = 12, fontweight = 'bold')

#Labels
plt.xlabel('Labor Force Rate', fontsize = 16)
plt.ylabel('Unemployment', fontsize = 16)

#Title
plt.title('Unemployment & Labor Force Rate correlation across NYC boroughs.', fontsize = 20, fontweight = 'bold')

#Show
plt.show()
```

<h4 style="color:#B7BA4A"><b> Negative Correlation: </b></h4>
<p style="color:#B7BA4A"> As the labor force increases, the unemployment dicreases.</p>
